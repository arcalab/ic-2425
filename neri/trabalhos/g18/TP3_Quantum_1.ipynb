### Question1 (GROVER ALG)

```python
#repetition (oracle + amplification once each)
#oracle
qc.x(0)
qc.x(2)

qc.h(2)
qc.ccx(0,1,2)
qc.h(2)


qc.x(0)
qc.x(2)


#amplification
qc.h(0)
qc.h(1)
qc.h(2)


qc.x(0)
qc.x(1)
qc.x(2)


qc.h(2)
qc.ccx(0,1,2)
qc.h(2)

qc.x(0)
qc.x(1)
qc.x(2)

qc.h(0)
qc.h(1)
qc.h(2)


qc.measure_all()

qc.draw()
```

### Question2 (NOISE)

```python
my_provider_ibmq = IBMQ.get_provider(hub='ibm-q', group='open', project='main') 
backend_device = my_provider_ibmq.get_backend('ibmqx2')
coupling_map = backend_device.configuration().coupling_map
```

```python
noise_model = NoiseModel.from_backend(backend_device)
print(noise_model)
```

```python
basis_gates = noise_model.basis_gates
```

```python
backend = Aer.get_backend("qasm_simulator")
result_noise = execute(qc, backend, 
                       noise_model=noise_model,
                       coupling_map=coupling_map,
                       basis_gates=basis_gates).result()

counts_noise = result_noise.get_counts(qc)
plot_histogram(counts_noise, title="Counts with depolarizing noise model")
```

```python
print(counts)
```

```python
print(counts_noise)
```

```python
def resume(counts_raw):
    s0=s1=0
    k=counts_raw.keys()
    lk=list(k)
    for c in lk:
        if c[0]=='0':
            s0 = s0 + counts_raw.get(c)
        else:
            s1 = s1 + counts_raw.get(c)
    return({'0':s0, '1':s1})
```

```python
cn = resume(counts_noise)
c = resume(counts)
```

```python
plot_histogram([c,cn], legend= ['simulation','simulation with noise'], color=['#061727','#82cfff'])
```
